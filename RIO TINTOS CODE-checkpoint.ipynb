```python
# Drop the location, not neccesary
df.drop(columns=['location','equipment_id'], inplace=True)
```

```python
#  categorical variables to numerical 
label_encoders = {}
for column in ['operation_mode', 'terrain_type']:  # Correcting the list of columns
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column])
    label_encoders[column] = le
    
    

```

```python
df.head() 
```

```python
# Feature Engineering before scaling
df['fuel_per_km'] = df['fuel_consumption_liters'] / (df['distance_traveled_km'] + 0.0001)
df['stress_index'] = df['temperature_celsius'] * df['humidity_percent'] / 100

# Verify the new features
print(df[['fuel_per_km', 'stress_index']].head())
```

```python
# Define the feature set and target variables
X = df.drop(columns=['maintenance_flag', 'downtime_hours', 'fuel_consumption_liters', 'timestamp'])
y_maintenance = df['maintenance_flag']  # For maintenance prediction
y_fuel = df['fuel_consumption_liters']  # For fuel efficiency prediction

# Split the dataset into training and testing sets for both tasks
X_train_maintenance, X_test_maintenance, y_train_maintenance, y_test_maintenance = train_test_split(X, y_maintenance, test_size=0.2, random_state=42)
X_train_fuel, X_test_fuel, y_train_fuel, y_test_fuel = train_test_split(X, y_fuel, test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()

# Include 'fuel_per_km' and 'stress_index' in the list of columns to scale
scaled_columns = ['load_weight_tonnes', 'distance_traveled_km', 'temperature_celsius', 'humidity_percent', 'fuel_per_km', 'stress_index']

# Scale the training data (fitting the scaler to training data only)
X_train_maintenance[scaled_columns] = scaler.fit_transform(X_train_maintenance[scaled_columns])
X_test_maintenance[scaled_columns] = scaler.transform(X_test_maintenance[scaled_columns])

# Scale the data for fuel prediction (similar process)
X_train_fuel[scaled_columns] = scaler.fit_transform(X_train_fuel[scaled_columns])
X_test_fuel[scaled_columns] = scaler.transform(X_test_fuel[scaled_columns])

# Verify scaling
print("First few rows of scaled training data for maintenance prediction:")
print(X_train_maintenance[scaled_columns].head())
```

```python
#EQUIPMENT MAINTANANCE


from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Calculate scale_pos_weight to handle class imbalance
scale_pos_weight = len(y_train_maintenance[y_train_maintenance == 0]) / len(y_train_maintenance[y_train_maintenance == 1])

# Initialize the Gradient Boosting model with adjusted scale_pos_weight
gbm_maintenance = XGBClassifier(random_state=42, scale_pos_weight=50)

# Train the model
gbm_maintenance.fit(X_train_maintenance, y_train_maintenance)

# Evaluate the model
y_pred_maintenance = gbm_maintenance.predict(X_test_maintenance)

# Print accuracy, precision, recall, and F1-score
print("Maintenance Model Accuracy:", accuracy_score(y_test_maintenance, y_pred_maintenance))
print(classification_report(y_test_maintenance, y_pred_maintenance))

# Optional: Print confusion matrix
print("Confusion Matrix:\n", confusion_matrix(y_test_maintenance, y_pred_maintenance))



```

```python
##FUEL EFFICIENCY

from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

# Initialize and train the Random Forest model for Fuel Efficiency
rf_fuel = RandomForestRegressor(random_state=42)
rf_fuel.fit(X_train_fuel, y_train_fuel)

# Evaluate the model
y_pred_fuel = rf_fuel.predict(X_test_fuel)
print("Fuel Model RMSE:", np.sqrt(mean_squared_error(y_test_fuel, y_pred_fuel)))
```

```python
rmse = np.sqrt(mean_squared_error(y_test_fuel, y_pred_fuel))
nrmse = rmse / (y_test_fuel.max() - y_test_fuel.min())
print("Relative RMSE:", nrmse)
```

```python
from sklearn.model_selection import GridSearchCV

# Hyperparameter tuning for GBM (Predictive Maintenance)
gbm_param_grid = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.05, 0.1],
    'max_depth': [3, 5, 7]
}
grid_gbm = GridSearchCV(estimator=gbm_maintenance, param_grid=gbm_param_grid, cv=5, scoring='accuracy')
grid_gbm.fit(X_train_maintenance, y_train_maintenance)

# Best parameters and model evaluation
best_gbm = grid_gbm.best_estimator_
y_pred_best_gbm = best_gbm.predict(X_test_maintenance)
print("Best GBM Maintenance Model Accuracy:", accuracy_score(y_test_maintenance, y_pred_best_gbm))
print(classification_report(y_test_maintenance, y_pred_best_gbm))

```

```python
# Hyperparameter tuning for Random Forest (Fuel Efficiency)
rf_param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10]
}
grid_rf = GridSearchCV(estimator=rf_fuel, param_grid=rf_param_grid, cv=5, scoring='neg_mean_squared_error')
grid_rf.fit(X_train_fuel, y_train_fuel)

# Best parameters and model evaluation
best_rf = grid_rf.best_estimator_
y_pred_best_rf = best_rf.predict(X_test_fuel)
print("Best RF Fuel Model RMSE:", np.sqrt(mean_squared_error(y_test_fuel, y_pred_best_rf)))
```

```python
from sklearn.ensemble import StackingClassifier, StackingRegressor

# Stacking Ensemble for Predictive Maintenance
estimators_maintenance = [('gbm', best_gbm)]
stacking_maintenance = StackingClassifier(estimators=estimators_maintenance, final_estimator=RandomForestClassifier())
stacking_maintenance.fit(X_train_maintenance, y_train_maintenance)
y_pred_stacking_maintenance = stacking_maintenance.predict(X_test_maintenance)
print("Stacking Maintenance Model Accuracy:", accuracy_score(y_test_maintenance, y_pred_stacking_maintenance))
print(classification_report(y_test_maintenance, y_pred_stacking_maintenance))


```

```python
# Stacking Ensemble for Fuel Efficiency
estimators_fuel = [('rf', best_rf)]
stacking_fuel = StackingRegressor(estimators=estimators_fuel, final_estimator=XGBRegressor())
stacking_fuel.fit(X_train_fuel, y_train_fuel)
y_pred_stacking_fuel = stacking_fuel.predict(X_test_fuel)
print("Stacking Fuel Model RMSE:", np.sqrt(mean_squared_error(y_test_fuel, y_pred_stacking_fuel)))
```
